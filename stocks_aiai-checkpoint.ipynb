```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_palette("Set2")
from datetime import datetime
import matplotlib.pyplot as plt
import math
```

```python
# IMPORT CSV FILE OF STOCK PRICES OVER A 5 YEAR PERIOD
df = pd.read_csv("stock_prices.csv")
df.head(5)
```

```python
# turn Date column into datetime using built in method
df['Date'] = pd.to_datetime(df['Date'])
df.head(5)
```

```python
# visualize price over time using seaborn lineplot
# use pd.melt to help format the table into long form for the hue
dfmelt = pd.melt(df, "Date")
dfmelt.head(5)
```

```python
sns.lineplot(data=dfmelt, x="Date", y="value", hue="variable")
plt.title("Securities over time")
plt.ylabel("Price ($)")
```

```python
# To get rid of exponential growth in some of our securities, make a new dataframe and take log of all columns
lndf = pd.DataFrame()
lndf["ln_APPLE"] = np.log(df["APPLE"])
lndf["ln_TESLA"] = np.log(df["TESLA"])
lndf["ln_BITCOIN"] = np.log(df["BITCOIN"])
lndf["ln_SPY"] = np.log(df["SPY"])
lndf["ln_GOLD"] = np.log(df["GOLD"])

lndf["Date"] = df["Date"]
lndf.head(5)
```

```python
# Once again, visualize this dataset
lndfmelt = pd.melt(lndf, "Date")
sns.lineplot(data=lndfmelt, x="Date", y="value", hue="variable")
plt.title("Securities over time")
plt.ylabel("Logged Price ($)")
```

```python
# Now, lets compare daily % returns using the original prices from our df dataframe. 
# To get the daily returns, we need to compare each day to the day before, and calculate the percent increase or decrease
# in price. The first row will be filled with np.nan, since there is no previous day to compare it to.
```

```python
pricechange = pd.DataFrame()
for column in df.columns.values:
    change = []
    for x in range(len(df) - 1):
        change.append(df[column][x+1] - df[column][x])
    columnlabel = column+" Price Change"
    pricechange[columnlabel] = change
```

```python
# delete date column and replace
pricechange["Date"] = pd.to_datetime(df["Date"])
pricechange = pricechange.drop(columns=["Date Price Change"])
pricechange
```

```python
# drop the last row of df and merge with percentreturn on date
df = df[:-1]
# add df's date column to our new pricechange table
pricechange["Date"] = df["Date"]
pricechange
```

```python
# now, merge on Date
mergeddf = df.merge(pricechange, how="outer")
mergeddf
```

```python
mergeddf
```

```python
weeklypercentchange = pd.DataFrame()
for column in df.drop(df.columns[0], axis=1).columns.values:
    change = []
    for x in range(len(df)):
        change.append(100*mergeddf[column+" Price Change"][x] / df[column][x])
    columnlabel = column+" % Change"
    weeklypercentchange[columnlabel] = change
weeklypercentchange["Date"] = mergeddf["Date"]
```

```python
# Once again, visualize this dataset
weeklypercentchangemelt = pd.melt(weeklypercentchange, "Date")
sns.lineplot(data=weeklypercentchangemelt, x="Date", y="value", hue="variable")
plt.title("Securities over time")
plt.ylabel("Logged Price ($)")
```

```python
# let's imagine we invested $100 into each one of these securities on our starting date, 2016-05-02. 
# How much money would we have today, for each of our securities? how much money would we have total?


# Build a new dataframe called "portfolio" that shows the value of your investment into those 5 securities over time
# This is your "equity" over time for each of these securities
# On 05-02-2016, your equity for each security should be $100.
portfolio = pd.DataFrame()
for column in df.drop(df.columns[0], axis=1).columns.values:
    price = [100]
    for x in np.arange(1,len(df)):
        price.append(weeklypercentchange[column+" % Change"][x] + price[x-1])
    columnlabel = column+" Equity"
    portfolio[columnlabel] = price
portfolio["Date"] = mergeddf["Date"]

```

```python
# Visualize your equity over time
portfolioemelt = pd.melt(portfolio, "Date")
sns.lineplot(data=portfolioemelt, x="Date", y="value", hue="variable")
plt.title("Securities over time")
plt.ylabel("Equity")
```

```python
# let's imagine we invested $100 into each one of these securities on our starting date, 2016-05-02. 
# According to our table, after 5 years we would have:
equityafter5years = portfolio[-1:]
apple = int(equityafter5years["APPLE Equity"])
tesla = int(equityafter5years["TESLA Equity"])
bitcoin = int(equityafter5years["BITCOIN Equity"])
spy = int(equityafter5years["SPY Equity"])
gold = int(equityafter5years["GOLD Equity"])
totalequity = apple+tesla+bitcoin+spy+gold

print("$"+str(apple)+" equity in Apple")
print("$"+str(tesla)+" equity in Tesla")
print("$"+str(bitcoin)+" equity in Bicoin")
print("$"+str(spy)+" equity in SPY")
print("$"+str(gold)+" equity in Gold")
print("$"+str(totalequity)+" total")
```

```python
# Though some of these securities did really well, some of them are considered very risky investments
# One measure of risk is called a stock's "volatility". Volatility is a statistical measure of the variance
# of the returns of a stock, and it's the most important determinant of how risky a security is.

# Here is the formula for volatility:

# annual volatility = std((log(current closing price / previous closing price)) * sqrt(trading weeks)
# Where "trading weeks" is 52

# Calculate the volatility of any stock

# Example of calculating Tesla's volatility using our original df dataframe:

# price at t:
tesla_price_t = np.array(df["TESLA"][1:])
tesla_price_t_minus_1 = np.array(df["TESLA"][:-1])
tesla_logged_returns = np.log(tesla_price_t / tesla_price_t_minus_1)

# NOTE: COULD ALSO HAVE DONE tesla_logged_returns = np.log(df['TESLA']/df['TESLA'].shift(1))

tesla_volatility = tesla_logged_returns.std() * (52**(1/2))
tesla_volatility
```

```python
# Do the same for SPY:

# price at t:
SPY_price_t = np.array(df["SPY"][1:])
SPY_price_t_minus_1 = np.array(df["SPY"][:-1])
SPY_logged_returns = np.log(SPY_price_t / SPY_price_t_minus_1)
SPY_volatility = SPY_logged_returns.std() * (52**(1/2))
SPY_volatility
```

```python
# Visualize the two securities' volatilities by plotting a histogram of their weekly logged returns on the same plot:
n_bins = 60
plt.hist(SPY_logged_returns, bins=n_bins, alpha=0.4, label="SPY")
plt.hist(tesla_logged_returns, bins=n_bins, alpha=0.4, label="TESLA")
plt.xlabel("Logged Weekly Returns")
plt.ylabel("Frequency")
plt.title("Weekly Logged Returns of Tesla and SPY")
plt.legend(loc='upper right')
plt.show()
```

```python
# This is an example of high risk high reward, vs low risk low reward.
# SPY is very clearly less volatile than TESLA during this time period, meaning the logged weekly returns are relatively low, regardless
# of whether they are gains or losses. This is consistent with our calculation for volatility of SPY and volatility of TESLA, since we concluded Tesla had a higher
# annual volatility than SPY. Why is this the case?
# Recall that SPY is an index-- it is not a single company's stock, but rather a compilation of dozens of hundreds of stocks. This appeals to people who want to 
# invest with low risk, because you're not putting all of your investment into one stock. By investing in SPY, you ensure that even if one company goes under, you won't
# lose all of your investment. Hence, it makes sense that the volatility of an index of hundreds of companies would be lower than the volatility of a single company like Tesla.
# It is ideal for long term investments, like retirement funds. On the other hand, notice how your $100 investment in Tesla turned into $472 after 5 years, whereas your 
# $100 SPY investment only turned into $179. 
# Conclusion:Reducing risk requires giving up potential gains (to save yourself from potential losses)
```
